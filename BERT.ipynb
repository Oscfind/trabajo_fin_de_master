## BERT

```python
import pandas as pd
from rouge import Rouge
from summarizer import Summarizer
```

```python
# Carga del dataset
df_ref = pd.read_csv("df_ref.csv")
df_ref
```

```python
# Crear un objeto Rouge para calcular la métrica ROUGE
rouge = Rouge()
```

```python
# Función para generar un resumen usando BERT
def generate_summary(text, num_sentences=3):
    summarizer = Summarizer()
    summary = summarizer(text, num_sentences=num_sentences)
    return summary
```

```python
# Suponiendo que df_nuevo es tu DataFrame con las columnas 'article' y 'abstract'
# Crear listas para almacenar los resultados de la métrica ROUGE
rouge_1_scores = []
rouge_2_scores = []
rouge_l_scores = []
```

```python
# Iterar sobre cada fila del DataFrame
for index, row in df_ref.iterrows():
    # Obtener el artículo y el resumen original
    article = row['article']
    original_summary = row['abstract']
    
    # Generar el resumen utilizando BERT
    generated_summary = generate_summary(article)
    
    # Calcular ROUGE para el resumen generado y el resumen original
    scores = rouge.get_scores(generated_summary, original_summary)
    
    # Extraer los valores específicos de ROUGE
    rouge_1_scores.append(scores[0]['rouge-1']['f'])
    rouge_2_scores.append(scores[0]['rouge-2']['f'])
    rouge_l_scores.append(scores[0]['rouge-l']['f'])
```

```python
# Crear un DataFrame para almacenar los resultados de ROUGE
rouge_scores_df = pd.DataFrame({
    'ROUGE-1': rouge_1_scores,
    'ROUGE-2': rouge_2_scores,
    'ROUGE-L': rouge_l_scores
})
```

```python
# Para posteriores cálculos se extrae el dataframe 'rouge_scores_BERT_df'
rouge_scores_df.to_csv('rouge_scores_BERT_df.csv', index=False)
```

```python
# Calcular estadísticas descriptivas si es necesario
# Por ejemplo, puedes calcular el promedio de las métricas ROUGE para todo el dataset
average_rouge_scores = rouge_scores_df.mean()
print("Promedio de métricas ROUGE para todo el dataset:")
print(average_rouge_scores)
```
