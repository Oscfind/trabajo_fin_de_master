```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

rouge_scores_TextRank_df = pd.read_csv("rouge_scores_TextRank_df.csv")
rouge_scores_LSA_df = pd.read_csv("rouge_scores_LSA_df.csv")
rouge_scores_KMEANS_df = pd.read_csv("rouge_scores_KMEANS_df.csv")
rouge_scores_T5_df = pd.read_csv("rouge_scores_T5_df.csv")
rouge_scores_BERT_df = pd.read_csv("rouge_scores_BERT_df.csv")
rouge_scores_BART_df = pd.read_csv("rouge_scores_BART_df.csv")

colors = sns.color_palette("pastel")

techniques = ['TextRank', 'LSA', 'K-MEANS', 'T5', 'BERT', 'BART']

def draw_boxplot(data, labels, title):
    plt.figure(figsize=(10, 6))
    bp = plt.boxplot(data, patch_artist=True)

    for i, box in enumerate(bp['boxes']):
        box.set(color='black', linewidth=2)
        box.set(facecolor=colors[i])

    plt.title(title)
    plt.ylabel('Score')

    for median in bp['medians']:
        median.set(color='red', linewidth=2)

    plt.xticks(range(1, len(labels) + 1), labels)

    plt.show()
```

```python
# Datos para ROUGE-1
data_rouge1 = [rouge_scores_TextRank_df['ROUGE-1'], rouge_scores_LSA_df['ROUGE-1'], rouge_scores_KMEANS_df['ROUGE-1'], rouge_scores_T5_df['ROUGE-1'], rouge_scores_BERT_df['ROUGE-1'], rouge_scores_BART_df['ROUGE-1']]
draw_boxplot(data_rouge1, techniques, 'ROUGE-1')
```

```python

# Datos para ROUGE-2
data_rouge2 = [rouge_scores_TextRank_df['ROUGE-2'], rouge_scores_LSA_df['ROUGE-2'], rouge_scores_KMEANS_df['ROUGE-2'], rouge_scores_T5_df['ROUGE-2'], rouge_scores_BERT_df['ROUGE-2'], rouge_scores_BART_df['ROUGE-2']]
draw_boxplot(data_rouge2, techniques, 'ROUGE-2')
```

```python
# Datos para ROUGE-L
data_rougeL = [rouge_scores_TextRank_df['ROUGE-L'], rouge_scores_LSA_df['ROUGE-L'], rouge_scores_KMEANS_df['ROUGE-L'], rouge_scores_T5_df['ROUGE-L'], rouge_scores_BERT_df['ROUGE-L'], rouge_scores_BART_df['ROUGE-L']]
draw_boxplot(data_rougeL, techniques, 'ROUGE-L')
```
