```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

BLEU_scores_TextRank_df = pd.read_csv("BLEU_scores_TextRank_df.csv")
BLEU_scores_LSA_df = pd.read_csv("BLEU_scores_LSA_df.csv")
BLEU_scores_Kmeans_df = pd.read_csv("BLEU_scores_Kmeans_df.csv")
BLEU_scores_T5_df = pd.read_csv("BLEU_scores_T5_df.csv")
BLEU_scores_BERT_df = pd.read_csv("BLEU_scores_BERT_df.csv")
BLEU_scores_BART_df = pd.read_csv("BLEU_scores_BART_df.csv")

colors = sns.color_palette("pastel")

techniques = ['TextRank', 'LSA', 'K-MEANS', 'T5', 'BERT', 'BART']

def draw_boxplot(data, labels, title):
    plt.figure(figsize=(10, 6))
    bp = plt.boxplot(data, patch_artist=True)

    for i, box in enumerate(bp['boxes']):
        box.set(color='black', linewidth=2)
        box.set(facecolor=colors[i])

    plt.title(title)
    plt.ylabel('Score')

    for median in bp['medians']:
        median.set(color='red', linewidth=2)

    plt.xticks(range(1, len(labels) + 1), labels)

    plt.show()
```

```python
# BLEU
data_bleu = [BLEU_scores_TextRank_df['BLEU'], BLEU_scores_LSA_df['BLEU'], BLEU_scores_Kmeans_df['BLEU'], BLEU_scores_T5_df['BLEU'], BLEU_scores_BERT_df['BLEU'], BLEU_scores_BART_df['BLEU']]
draw_boxplot(data_bleu, techniques, 'BLEU')
```
